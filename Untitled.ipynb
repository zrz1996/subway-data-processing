```python
from igraph import *
```

```python
inputfile = open('graph.txt', 'r')
```

```python
edges = []
```

```python
for line in inputfile:
    edges.append([int(x) for x in line.split()])
inputfile.close()
```

```python
edges
```

```python
g = Graph(edges)
```

```python
g.delete_vertices(0)
```

```python
summary(g)
```

```python
adj = g.get_adjacency()
```

```python
N = 300
```

```python
g.es[g.incident(2)]
```

```python
cur_edges = []
def dfs(x, rem):
    if rem == 0:
        return
    for i in range(N):
        if adj[x - 1][i] > 0:
            cur_edges.append((x, i + 1))
            adj[x - 1][i] = 0
            adj[i][x - 1] = 0
            dfs(i + 1, rem - 1)
            adj[x - 1][i] = 1
            adj[i][x - 1] = 1
```

```python
degree = []
for v in range(N):
    print(v)
    cur_edges = []
    dfs(v + 1, 9)
    cur_edges = sorted(set(cur_edges))
    degree.append(cur_edges)
```

```python
d = [len(degree[i]) for i in range(N)]
```

```python
f = open('degree.txt', 'w')
```

```python
for i in range(N):
    f.write("%s\n" % d[i])
```

```python
f.close()
```

```python
f = open('links.txt', 'w')
for i in range(N):
    f.write("%s:\n" % (i + 1))
    for item in degree[i]:
        f.write("(%s, %s) " % item)
    f.write("\n")
f.close()
```

```python
for i in range(N):
    for j in range(len(degree[i])):
        degree[i][j] = sorted(set(degree[i][j]))
```

```python
for i in range(N):
    for j in range(len(degree[i])):
        degree[i][j] = tuple(degree[i][j])
    degree[i] = list(set(degree[i]))
```

```python
for i in range(N):
    degree[i] = sorted(degree[i])
```
